# Data Selection 

```python
# lets change the index first to understand it better
# This will change index of our DataFrame from RangeIndex to
# 'Ticket' values
titanic_ticket_index = titanic_data.set_index('Ticket')
```

```python
# checking records, see index has been changed to ticket
titanic_ticket_index.head(5)
```

## loc- select data based on labels

```python
# select single row with value A/5 21171
titanic_ticket_index.loc['A/5 21171']
```

```python
# select multiple rows with labels method 1: using list
row_labels = ['A/5 21171', 'PC 17599', '373450']
titanic_ticket_index.loc[row_labels]
```

```python
# select multiple rows with labels method 2: using ':'

# we can supply start_label:end_label like
# here we are selecting rows from label 'PC 17599' to '373450'
# Note- this will not work if there are multiple rows with the same labels 
titanic_ticket_index.loc['PC 17599':'373450']
```

```python
# select single column Embarked
titanic_ticket_index.loc[:,'Embarked'].head(5)
```

```python
# Selecting multiple columns: Sex, Age, Fare, Embarked column for all rows.
titanic_ticket_index.loc[:,['Sex','Age','Fare','Embarked']].head(5)
```

```python
# Selecting multiple columns using ':' method 1
# selecting columns from starting and until 'Age column'
titanic_ticket_index.loc[:, :'Age'].head(5)
```

```python
# Selecting multiple columns using ':' method 2
# selecting columns from 'Age column' to the end
titanic_ticket_index.loc[:, 'Age':].head(5)
```

```python
# select multiple columns with labels using ':' method 3

# we can supply column start_label:end_label like
# here we are selecting columns from label 'Sex' to 'Embarked'
titanic_ticket_index.loc[:, 'Sex':'Embarked'].head(5)
```

```python
# selecting specific rows and columns
titanic_ticket_index.loc[['A/5 21171','PC 17599', '373450'], ['Sex','Fare','Age','Embarked']]
```

## iloc- selecting data based on index

Now we will try to get the same results using iloc which we got using loc

```python
# selecting first row
titanic_ticket_index.iloc[0]
```

```python
# select multiple rows, same rows as what we selected in loc
titanic_ticket_index.iloc[[0, 1, 4]]
```

```python
# select multiple rows, using ':'
# we can supply start_index:end_index 
# in case of selecting from starting we can omit the start_index
titanic_ticket_index.iloc[1:5]
```

```python
# select Embarked column which is at index 10
titanic_ticket_index.iloc[:, 10].head(5)
```

```python
# Selecting multiple columns: Sex, Age, Fare, Embarked column for all rows.
titanic_ticket_index.iloc[:,[4, 5, 8, 10]].head(5)
```

```python
# Selecting multiple columns using ':' method 1
# selecting columns from starting and until 'Age column'
titanic_ticket_index.iloc[:, :6].head(5)
```

```python
# Selecting multiple columns using ':' method 2
# selecting columns from 'Age column' to the end
titanic_ticket_index.iloc[:, 5:].head(5)
```

```python
# Selecting multiple columns using ':' method 3

# we can add start_index:end_index for columns
# selecting 1st column to the 5th column
titanic_ticket_index.iloc[:, 1:5].head(5)
```

```python
# selecting specific rows and columns: example 1
titanic_ticket_index.iloc[[0, 1, 4], [4, 5, 8, 10]]
```

```python
# selecting specific rows and columns: example 2
# we can use start_index:end_index for both columns and rows.

# selecting 3rd to 6th row and 1st to 4th column (right limit is not included)
titanic_ticket_index.iloc[3:7, 1:5]
```

# Filtering the data

```python
# first 5 records that has Age > 15
titanic_data.query('Age > 15').head(5)
```

```python
# first 5 males who survived
titanic_data.query('Sex=="male" and Survived==1').head(5)
```

```python
# using user defined variables

# gender to select and min_fare, these can be passed as part of argument to the Script
gender_to_select = "female"
min_fare = 50

# querying using attribute passed
titanic_data.query('(Sex==@gender_to_select) and (Fare > @min_fare)').head(5)
```

```python
# Age > 15 and were travelling with a Sibling or Parent
titanic_data.query('Age > 15 & (SibSp==1 | Parch==1)').head(5)
```

# Saving Data

```python
# using to_csv

# lets save the data for the males who survived
# we can use chaining to first filter and then save
titanic_data.query('Sex=="male" and Survived==1').to_csv('males_survived.csv', index=False)
```

```python
# using to_excel

# lets save the data for passengers travelling with a Sibling or Parent
# we can use chaining to first filter and then save
titanic_data.query('SibSp==1 | Parch==1').to_excel('travelling_with_parent_sibling.xlsx', index=False)
```
